#### Problem Statement

The goal of this task is to explore the Iris dataset and understand how different features relate to flower species.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data():
    # Using seaborn's built-in dataset
    df = sns.load_dataset("iris")
    return df
```

```python
def explore_data(df):
    # understanding  what we're working with
    print("First 5 rows of the dataset:")
    print(df.head())
    
    print("\nShape of dataset:", df.shape)

    # Basic statistics for distributions
    print("\nStatistical Summary:")
    print(df.describe())
    
    # Checking data types and missing values
    print("\nDataset Info:")
    print(df.info())
    
    
```

```python
def visualize_data(df):
    # Trying to see if different species separate visually using sepal dimensions
    sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue = "species")
    plt.title("Sepal Length vs Sepal Width")            
    plt.xlabel("Sepal Length")
    plt.ylabel("Sepal Width")
    plt.show()

    # Looking at how sepal length is distributed across species
    sns.histplot(data=df, x="sepal_length", hue = "species", kde=True)
    plt.title("Distribution of Sepal Length by Species")    
    plt.xlabel("Sepal Length")
    plt.ylabel("Frequency")
    plt.show()

    # Boxplot to check spread and possible outliers in sepal length
    sns.boxplot(x="species",y="sepal_length",data=df,palette="Set2",hue="species",legend=False)
    plt.title("Sepal Length Distribution by Species")
    plt.xlabel("Species")
    plt.ylabel("Sepal Length")
    plt.show()
```

```python
# Step 1: Load dataset
df = load_data()

# Step 2: Explore dataset
explore_data(df)

# Step 3: Visualize dataset
visualize_data(df)
```
